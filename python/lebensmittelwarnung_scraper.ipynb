```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.lebensmittelwarnung.de/bvl-lmw-de/liste/alle/bremen/150/0"
response = requests.get(url)
print(response)

soup = BeautifulSoup(response.content, 'html.parser')
```

```python
data_containers = soup.find_all('div', class_='lmw-warnungsliste')
# print(data_containers[0])

# Initialize lists to store data
data = {
    'product': [],
    'manufacturer': [],
    'reason': [],
    'address': [],
    'affected' : []
}

# Loop through each data container
for id, container in enumerate(data_containers):
    # Find the span element with the specified id

    product = container.find('span', class_='preWrap').text.strip()
    manufacturer_data = container.find('div', class_='hersteller').text.strip()
    manufacturer = manufacturer_data.split('\n')[2]
    address = ''.join(manufacturer_data.split('\n')[3:]).replace('\r', ' ')
    reason = container.find('div', class_='grundderwarnung').text.strip().split('\n')[2]
    affected = container.find('div', class_='betroffeneLaender').text.strip()
    affected = affected.split("\n")[2]

    data['product'].append(product)
    data['manufacturer'].append(manufacturer)
    data['reason'].append(reason)
    data['address'].append(address)
    data['affected'].append(affected)

df = pd.DataFrame(data)
df
```
