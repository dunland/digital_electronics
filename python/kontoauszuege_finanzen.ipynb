```python
import pandas as pd

df = pd.read_csv("/home/dav/Documents/GLS/Kontoauszuege/Umsaetze_DE80430609674100579900_2023.12.13.csv", delimiter=';', decimal=',')

df['Bemerkung'] = df['Bemerkung'].replace("miete", "Miete")
df['Bemerkung'] = df['Bemerkung'].replace("urlaub", "Urlaub")
df['Bemerkung'] = df['Bemerkung'].replace("alltag", "Alltag")
df['Bemerkung'] = df['Bemerkung'].replace("tech", "Tech")
df['Bemerkung'] = df['Bemerkung'].replace("Bankgebühr", "Alltag")
df['Bemerkung'] = df['Bemerkung'].replace("gönn", "Hobby")

df['Bemerkung'].fillna('Sonstiges', inplace=True)
df['Buchungstag'] = pd.to_datetime(df['Buchungstag'], format="%d.%m.%Y")
df = df.sort_values(by='Buchungstag', ascending=True).reset_index(drop=True)

print(df.columns)

df[df['Name Zahlungsbeteiligter'] == 'Landschaftliche Brandkasse Hannover']
```

```python
# Automatisiere:
df1 = df

def update_Bemerkung(filter_column,filter_keyword,bemerkung):
    for idx, row in enumerate(df[filter_column]):
        if type(row) == str:
            if row.__contains__(filter_keyword):
                df1.loc[idx, 'Bemerkung'] = bemerkung
                print(row)
    df.update(df1)

update_Bemerkung("Verwendungszweck", "Spende", "Spende")
update_Bemerkung("Verwendungszweck", "Abschluss per", "Bankgebühr")
update_Bemerkung("Verwendungszweck", "Klara, Theda", "Spende")
update_Bemerkung("Verwendungszweck", "Miete", "Miete")
update_Bemerkung("Verwendungszweck", "Proberaummiete", "Miete")
update_Bemerkung("Verwendungszweck", "VERSICHERUNG", "Versicherung")
update_Bemerkung("Name Zahlungsbeteiligter", "MISEREOR", "Spende")
update_Bemerkung("Name Zahlungsbeteiligter", "betterplace.org", "Spende")
update_Bemerkung("Name Zahlungsbeteiligter", "Drillisch", "Telefon")
update_Bemerkung("Buchungstext", "Lohn/Gehalt/Rente", "Lohn")
```

```python
# Exportiere Datei:
df.to_csv("/home/dav/Documents/GLS/Kontoauszuege/Umsaetze_DE80430609674100579900_2023.12.13_export.csv", sep=";", index=False)
```
